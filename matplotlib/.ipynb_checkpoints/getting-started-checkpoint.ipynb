```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
xaxis=np.arange(1,9)
yaxis=np.arange(1,31,4)

plt.plot(xaxis, yaxis)
plt.show()

plt.plot(xaxis, yaxis,color='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("First Python Plot")
plt.show()
```

```python
plotx=np.linspace(3,-3,7)
ploty=plotx**2
plt.plot(plotx, ploty)
plt.show()
```

```python
plotx=np.linspace(4,-4,50)
ploty=plotx**2
plt.plot(plotx, np.sin(plotx),'r')
plt.plot(plotx, np.cos(plotx),'g')
plt.show()
```

```python
pltx=np.arange(1,11)
plty=np.arange(10,101,10)

plt.plot(pltx, plty)
plt.show()

plt.bar(pltx, plty)
plt.barh(pltx, plty)
plt.polar(pltx, plty)
plt.step(pltx, plty)
```

```python
plotsx=np.arange(1,10)
plotsy=np.arange(3,30,3)

plt.plot(plotsx, plotsy)
plt.subplot(2,2,1)
plt.plot(plotsx, plotsy, color='green')
plt.subplot(2,2,2)
plt.plot(plotsx, plotsy, color='red')
plt.subplot(2,2,3)
plt.plot(plotsx, plotsy, color='blue')
plt.subplot(2,2,4)
plt.plot(plotsx, plotsy, color='black')
```

```python
plt.subplot(2,2,1)
plt.plot(plotsx, plotsx**2, color='green')
plt.subplot(2,2,2)
plt.plot(plotsx, np.log10(plotsx), color='red')
plt.subplot(2,2,3)
plt.plot(plotsx, np.exp(plotsx), color='blue')
plt.subplot(2,2,4)
plt.plot(plotsx, np.sqrt(plotsx), color='black')
```

```python
Figure=plt.figure()

axis=Figure.add_axes([0,0,1,1])

axis.plot(plotsx, plotsy, color='red')
```

```python
axis.set_xlabel('X values')
axis.set_ylabel('Y values')
Figure
```

```python
axis.legend(labels='Label', loc='lower right')
Figure
```

```python
x1=np.arange(1,11)
y1=np.arange(3,31,3)

Figure, Axes=plt.subplots(nrows=2,ncols=2)
plt.tight_layout()
```

```python
Figure, Axes=plt.subplots(nrows=2,ncols=1)
Axes[0].plot(x1,y1,color='red')
Axes[0].set_title('Line Garph')
Axes[1].plot(x1,x1**2,color='red')
plt.tight_layout()

Figure, Axes=plt.subplots(nrows=2,ncols=1)
Axes[0].plot(x1,np.sqrt(x1),color='green')
Axes[1].plot(x1,np.exp(x1),color='green')
plt.tight_layout()

Figure, Axes=plt.subplots(nrows=2,ncols=1)
Axes[0].plot(x1,np.sin(x1),color='black')
Axes[1].plot(x1,np.log10(x1),color='black')
plt.tight_layout()

Figure, Axes=plt.subplots(nrows=2,ncols=1)
Axes[0].plot(x1,np.sin(x1),y1,np.sin(y1),color='blue')
Axes[1].plot(x1,np.tan(x1),y1,np.tan(y1),color='blue')
plt.tight_layout()
```

```python
Figures=plt.figure()

axes1=Figures.add_axes([0,0,1,1])

axes1.plot(x1,np.sqrt(y1),color='red')
```

```python
Figurex, Axesx=plt.subplots(figsize=(7,7))
Axesx=Figurex.add_axes([0,0,1,1])

Axesx.plot(x1,np.sqrt(x1),label='X-sqrt',color='green')
Axesx.plot(x1,x1**2,label='X^2',color='red')
Axesx.plot(y1,np.sqrt(y1),label='Y-sqrt',color='blue')
Axesx.plot(y1,y1**2,label='Y^2',color='black')
Axesx.legend()
plt.tight_layout()
```
